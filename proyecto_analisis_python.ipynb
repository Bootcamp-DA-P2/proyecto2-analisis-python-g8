# Importación de dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Exploración inicial

```python
df_kiva_loans= pd.read_csv('/content/kiva_loans_def.csv') #Lectura de datos
df_kiva_loans.head #Muestra de filas por defecto
df_kiva_loans.tail(10000) #Muestra de datos
df_kiva_loans.shape #Devuelve una tupla a dimensiones del dataset
df_kiva_loans.describe() #Generación de estadísticas
```

# Valores nulo

```python
df_kiva_loans = pd.read_csv('/content/kiva_loans_def.csv')
df_kiva_loans ['Prueba_Null'] = None
display(df_kiva_loans.head()) #identificar nulos
df_kiva_loans.isnull()
df_kiva_loans.fillna('faltante', inplace=True) #reemplazar valores nulos con la palabra "faltante"
df_kiva_loans.replace('faltante', '', inplace=True) #reemplazar los valores por comillas ''
df_kiva_loans.info()
```

# Manejar valores nulos

```python
df_kiva_loans.dropna() #Elimina filas que contenga un valor nulo
df_kiva_loans.dropna(how='all') #Elimina fila si sus valores son nulos
df_kiva_loans.dropna(subset=['tags']) #Elimina columna "tags"
df_kiva_loans.drop_duplicates(inplace=True) #Eliminación para evitar sobre-representación
df_kiva_loans
```

# Normalizar texto

```python
df_kiva_loans['borrewer_genders'] = df_kiva_loans['borrower_genders'].str.strip() #eliminación de espacios blancos accidentales
cols_texto = df_kiva_loans.select_dtypes(include=['object']).columns #identifica las columnas de formato texto
df_kiva_loans[cols_texto].apply(lambda x: x.str.strip().str.lower()) #Aplica limpieza con minúsculas
```

#Formatos y fechas

```python
date_cols = ['posted_time', 'disbursed_time', 'funded_time'] #Corregir formato fecha
for col in date_cols:
    df_kiva_loans[col] = pd.to_datetime(df_kiva_loans[col], errors='coerce', utc=True)
df_kiva_loans

```

#Formato monetario

```python
df_kiva_loans['loan_amount_formatted'] = df_kiva_loans['loan_amount'].apply(lambda x: f"€{x:,.2f}") #cambio de int a string
df_kiva_loans
```

#Formato numérico

```python
df_kiva_loans['loan_amount'] = pd.to_numeric(df_kiva_loans['loan_amount'], errors='coerce') #Corregir formato a número en columnas específicas
df_kiva_loans['funded_amount'] = pd.to_numeric(df_kiva_loans['funded_amount'], errors='coerce')
df_kiva_loans
```

#Creación de Columnas

```python
limites = [25.0, 50000.0, 100000.0, float('inf')] #añadir columnas variable contigua (loan_amount)
nombres_categorias = ['bajo', 'medio', 'alto']
df_kiva_loans['rango_prestamo'] = pd.cut(df_kiva_loans['loan_amount'],
                                         bins=limites,
                                         labels=nombres_categorias)
df_kiva_loans.sort_values("loan_amount", ascending= True) #ordenar ascendente los valores
df_kiva_loans.sort_values("borrower_genders", ascending=False) #ordenar de manera descendente
conteo_generos = df_kiva_loans['term_in_months'].value_counts() #conteo de plazos
df_kiva_loans
```

#Outliers

```python
Q1 = df_kiva_loans['loan_amount'].quantile(0.25) #Calcula los cuartiles
Q3 = df_kiva_loans['loan_amount'].quantile(0.75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR #Definición de límites
limite_superior = Q3 + 1.5 * IQR
outliers = df_kiva_loans[(df_kiva_loans['loan_amount']< limite_inferior) | (df_kiva_loans['loan_amount'] > limite_superior)] #Filtración y muestra de Outliers
print(f"Se encontraron {len(outliers)}outliers.")
print(outliers[['loan_amount', 'country', 'term_in_months' ]])
```

#Visualización de datos

```python
top_10_paises = df_kiva_loans['country'].value_counts().index[:10] #Creación de ranking con los 10 paises
df_top = df_kiva_loans[df_kiva_loans['country'].isin(top_10_paises)]

plt.figure(figsize=(12,6))
sns.countplot(data=df_top, x= 'country', hue='rango_prestamo', palette='magma')

plt.title('Distribucion de Préstamos en Ranking Top 10 Paises')
plt.xlabel('Pais')
plt.ylabel('Cantidad de Préstamos')
plt.xticks(rotation=45)
plt.legend(title='Rango')
plt.show()
```

```python
top_10_paises = df_kiva_loans['region'].value_counts().index[:10] #Creación de ranking con los 10 regiones
df_top = df_kiva_loans[df_kiva_loans['region'].isin(top_10_paises)]

plt.figure(figsize=(12,6))
sns.countplot(data=df_top, x= 'region', hue='rango_prestamo', palette='magma')

plt.title('Distribucion de Préstamos en Ranking Top 10 Regiones')
plt.xlabel('Region')
plt.ylabel('Cantidad de Préstamos')
plt.xticks(rotation=45)
plt.legend(title='Rango')
plt.show()
```
